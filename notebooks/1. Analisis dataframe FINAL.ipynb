## 1. ANÁLISIS DATAFRAME FINAL

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib as plt
import matplotlib.pyplot as plt


import statsmodels.formula.api as smf
```

Features explanation of the  dataset:

    + id: only for test & sample submission files, id for prediction sample identification
    + price: price in USD
    + carat: weight of the diamond
    + cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
    + color: diamond colour
    + clarity: a measurement of how clear the diamond is
    + x: length in mm
    + y: width in mm
    + z: depth in mm
    + depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
    + table: width of top of diamond relative to widest point (43--95)

### 1.1 Análisis de las variables categóricas

    1.1.2 CLARITY
    1.1.3 COLOR

##### 1.1.1 CUT

Vamos a ordenar numéricamente el corte de los diamantes siendo 5 el mejor y 1 el peor corte

```python
data_train.cut.unique()
```

```python
cut_map = {
    "Ideal": 5,
    "Premium": 4,
    "Very Good": 3,
    "Good": 2,
    "Fair": 1 
}
```

```python
data_train["cut_num"] = data_train.cut.map(cut_map)
data_train
```

##### 1.1.2 CLARITY
Vamos a calcular el precio medio por la claridad del diamante para averiguar cuales son los precios diamantes dependiendo de su claridad

```python
data_train.clarity.nunique()
```

```python
data_train.head()
```

##### 1.1.3 COLOR

Con la columna de color lo que vamos a realizar va a ser: implentar la función "dummies" para crear una columna numérica que nos pueda servir para la aplicación de modelos estadísticos y poder estudiar el precio del diamante según sea su color. 
Como la columna de color no muestra nigún tipo de orden entre las caracteísticas que la forman implementamos esta función para la facilidad en su futuro estudio estadístico.

```python
data_train.color.unique()
```

```python
data_train = pd.get_dummies(data_train, columns=["color"], drop_first=True)
```

```python
data_train.isnull().sum()
```

```python
data_train.head()
```

```python
# Por ultimo eliminamos la columna "cut" para tener nuestra columnas numericas
data_train_clean = data_train.drop("cut", axis=1)
data_train_clean
```

```python
# Comprobamos que nuestro dataset este formado solamente por columnas numéricas
data_train_clean.info()
```

```python
# GUARDAMOS NUESTRO DATAFRAME LIMPIO EN LA CARPETA DE DATA
data_train_clean.to_csv("../data/data_train_clean.csv")
```

### 1.2 COLIANIDAD ENTRE LAS VARIABLES

```python
data_train_clean.corr()
```

```python
results = smf.ols('price ~ carat + depth + x + y + z + table + cut_num + clarity_IF + clarity_SI1 + clarity_SI2 + clarity_VS1 + clarity_VS2 + clarity_VVS1 + clarity_VVS2 + color_E + color_F + color_G + color_H + color_H + color_I + color_J', data=data_train_clean).fit()
results.summary()
```

```python
results = smf.ols('price ~ clarity_VS1 + clarity_VS2 + clarity_VVS2+ clarity_VVS1+ clarity_IF + clarity_SI2 + table + x', data=data_train_clean).fit()
results.summary()
```

* Lo que podemos observar es que la claridad del diamante tiene muchsísima correlación con su precio
* Y la X (lenght in mm) tambien tiene una fuerte correlación con el precio del diamante

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(data_train_clean.corr(),center=0,  cmap='Blues', annot=True)
```

* En esta gráfica podemos ver muchísimo más visual las correlaciones entre las variables
* Podemos observar que las mayores colinealidades que existen entre el presio son con la columna "x, y, z y carat"
* Además pordemos observar el resto de las variables que tienen correlacion entre el precio y por ello abajo intentamos crear un modelo con esas pariables obteniendo un r_squared ajustado de 95.5 %
* El siguiente modelo no es el perfecto (es el anterior al perfecto) pero, estamos a punto de dar con él 

```python
results = smf.ols('price ~ x + y + z + carat + depth + table + clarity_VS1 +clarity_VS2+ clarity_SI1 + clarity_SI2 + color_F + color_G + color_H + color_H + color_I + color_J', data=data_train_clean).fit()
results.summary()
```
